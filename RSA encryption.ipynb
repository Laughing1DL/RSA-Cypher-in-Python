```python
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP
import base64

# Generate keys
key = RSA.generate(2048)
public_key = key.publickey().exportKey()
private_key = key.exportKey()

# Encryption
message = b"Secure communication with RSA"
cipher_rsa = PKCS1_OAEP.new(RSA.importKey(public_key))
ciphertext = cipher_rsa.encrypt(message)
print(f"Ciphertext: {base64.b64encode(ciphertext).decode()}")

# Decryption
decipher_rsa = PKCS1_OAEP.new(RSA.importKey(private_key))
decrypted_text = decipher_rsa.decrypt(ciphertext)
print(f"Decrypted text: {decrypted_text.decode()}")
```
